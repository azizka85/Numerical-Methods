### Example

$y = x \cdot ln(x)$, where $x \in [1, 10]$

##### Calculating $F(x) = \int_1^x f(x)dx$ using Newton-Cotes formulas

```python
import numpy as np
from math import log
import modules.integration.newton_cotes as nc

a = 1
b = 10
k = 3

def f(x): return x * log(x)

dx = [1, 0.1, 0.01]

x = [np.arange(a, b + d / 2, d) for d in dx]

i_nc = [np.zeros(len(t)) for t in x]

for i in range(len(x)):
    n = len(x[i])

    for j in range(n):
        i_nc[i][j] = nc.integral(f, a, x[i][j], k)        
```

```python
from matplotlib import pyplot as plt

for i in range(len(dx)):
    plt.plot(x[i], i_nc[i], label=f'dx = {dx[i]}')

plt.title("Calculating using Newton-Cotes formulas")
plt.xlabel('x')
plt.grid()
plt.legend()
```

##### Calculating $F(x) = \int_1^x f(x)dx$ using Gaussian quadratures

```python
import modules.integration.gauss as g

i_g = [np.zeros(len(t)) for t in x]

for i in range(len(x)):
    n = len(x[i])

    for j in range(n):
        i_g[i][j] = g.integral(f, a, x[i][j], k)   
```

```python
for i in range(len(dx)):
    plt.plot(x[i], i_g[i], label=f'dx = {dx[i]}')

plt.title("Calculating using Gaussian quadratures")
plt.xlabel('x')
plt.grid()
plt.legend()
```

##### Comparison $F(x) = \int_1^x f(x)dx$ for $dx = 1$

```python
plt.plot(x[0], i_nc[0], label=f'Newton-Cotes')
plt.plot(x[0], i_g[0], label=f'Gauss')

plt.title("Comparison for dx=1")
plt.xlabel('x')
plt.grid()
plt.legend()
```
