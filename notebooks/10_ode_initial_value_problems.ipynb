### Пример №1

$y'' = -0.1y' - x$, $y(0) = 0$, $y'(0) = 1$

$y' = z \Rightarrow \begin{bmatrix}
y' \\
z' \\
\end{bmatrix} = \begin{bmatrix}
z \\
-0.1z - x \\
\end{bmatrix} \Rightarrow f(x, y, z) = \begin{bmatrix}
z \\
-0.1z - x \\
\end{bmatrix}$ 

##### Нахождение $y = y(x)$ и $z = y'(x)$ методом Эйлера

```python
import numpy as np
from modules.ode.linear.initial_value import euler

def f(x: float, y: np.ndarray[float]) -> np.ndarray[float]:
    fx = np.zeros(len(y))
    
    fx[0] = y[1]
    fx[1] = -0.1*y[1] - x

    return fx

a = 1.
b = 3.

ya = np.array([0., 1.])

N = [5, 25, 125]

re = [euler(f, a, b, ya, n) for n in N]
```

```python
import matplotlib.pyplot as plt

for i in range(len(N)):
    plt.plot(re[i][0], re[i][1][:, 0], label=f'N = {N[i]}')

plt.title("Решение y = y(x) методом Эйлера")
plt.xlabel('x')
plt.grid()
plt.legend()
```

```python
for i in range(len(N)):
    plt.plot(re[i][0], re[i][1][:, 1], label=f'N = {N[i]}')

plt.title("Решение z = y'(x) методом Эйлера")
plt.xlabel('x')
plt.grid()
plt.legend()
```

##### Нахождение $y = y(x)$ и $z = y'(x)$ модифицированным методом Эйлера

```python
from modules.ode.linear.initial_value import modified_euler

rme = [modified_euler(f, a, b, ya, n) for n in N]
```

```python
for i in range(len(N)):
    plt.plot(rme[i][0], rme[i][1][:, 0], label=f'N = {N[i]}')

plt.title("Решение y = y(x) модифицированным методом Эйлера")
plt.xlabel('x')
plt.grid()
plt.legend()
```

```python
for i in range(len(N)):
    plt.plot(rme[i][0], rme[i][1][:, 1], label=f'N = {N[i]}')

plt.title("Решение z = y'(x)")
plt.xlabel('x')
plt.grid()
plt.legend()
```

```python
plt.plot(re[1][0], re[1][1][:, 0], label='Метод Эйлера')
plt.plot(rme[1][0], rme[1][1][:, 0], label='Модифицированный метод Эйлера')

plt.title(f"Сравнение y = y(x), N = {N[1]}")
plt.xlabel('x')
plt.grid()
plt.legend()
```

### Пример №2

$y'' = -1001y' - 1000y$, $y(0) = 0$, $y'(0) = 1$

$y' = z \Rightarrow \begin{bmatrix}
y' \\
z' \\
\end{bmatrix} = \begin{bmatrix}
z \\
-1001z - 1000y \\
\end{bmatrix} \Rightarrow f(x, y, z) = \begin{bmatrix}
z \\
-1001z - 1000y \\
\end{bmatrix}$ 

##### Нахождение $y = y(x)$ методом Эйлера

```python
def f(x: float, y: np.ndarray[float]) -> np.ndarray[float]:
    fx = np.zeros(len(y))
    
    fx[0] = y[1]
    fx[1] = -1001*y[1] - 1000*y[0]

    return fx

a = 0.
b = 2.
n = 25

ya = np.array([0., 1.])

re = euler(f, a, b, ya, n)
```

```python
plt.plot(re[0], re[1][:, 0])
plt.title(f"Решение y = y(x) методом Эйлера, N = {n}")
plt.xlabel('x')
plt.grid()
plt.legend()
```

```python
plt.plot(re[0], re[1][:, 1])
plt.title(f"Решение z = y'(x) методом Эйлера, N = {n}")
plt.xlabel('x')
plt.grid()
plt.legend()
```

##### Нахождение $y = y(x)$ модифицированным методом Эйлера

```python
rme = modified_euler(f, a, b, ya, n)
```

```python
plt.plot(rme[0], rme[1][:, 0])
plt.title(f"Решение y = y(x) модифицированным методом Эйлера, N = {n}")
plt.xlabel('x')
plt.grid()
plt.legend()
```

```python
plt.plot(rme[0], rme[1][:, 1])
plt.title(f"Решение z = y'(x) модифицированным методом Эйлера, N = {n}")
plt.xlabel('x')
plt.grid()
plt.legend()
```

##### Нахождение $y = y(x)$ и $z = y'(x)$ неявным методом Эйлера

$y'' = -1001y' - 1000y$, $y(0) = 0$, $y'(0) = 1$

$
\begin{cases}
    y' = z \\
    z' = -1001z - 1000y \\
\end{cases} \Rightarrow
$
$
\begin{cases}
    y_{i+1} - y_i = hz_{i+1} \\
    z_{i+1} - z_i = -1001hz_{i+1} - 1000hy_{i+1} \\
\end{cases} \Rightarrow
$
$
\begin{cases}
    y_{i+1} - hz_{i+1} = y_i \\
    (1 + 1001h)z_{i+1} + 1000hy_{i+1} = z_i \\
\end{cases}
$

```python
from modules.slae.solver.direct.gauss_elimination import gauss_elimination

m = len(ya)
    
x = np.linspace(a, b, n)
y = np.zeros((n, m))

y[0] = ya
h = x[1] - x[0]

a = np.array([
    [1., -h],
    [1000.*h, 1. + 1001.*h]
])

for i in range(1, n):    
    y[i] = gauss_elimination(a.copy(), y[i-1].copy())
```

```python
plt.plot(x, y[:, 0])
plt.title(f"Решение y = y(x) неявным методом Эйлера, N = {n}")
plt.xlabel('x')
plt.grid()
plt.legend()
```

```python
plt.plot(x, y[:, 1])
plt.title(f"Решение z = y'(x) неявным методом Эйлера, N = {n}")
plt.xlabel('x')
plt.grid()
plt.legend()
```
