### Пример

$y = x \cdot ln(x)$, где $x \in [1, 10]$

```python
import numpy as np
import random as rnd
from modules.fitting.poly_fit import poly_fit, std_dev, plot_poly

xd = np.arange(1., 10., 0.1)
yd = xd * np.log(xd)

for i in range(len(yd)):
    yd[i] = yd[i] + yd[i] * rnd.random() * 0.1

c = poly_fit(xd, yd, 3)

plot_poly(xd, yd, c)
```

```python
std_dev(c, xd, yd)
```
