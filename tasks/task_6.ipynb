### 1. Метод последовательного поиска

##### Пример:

$$x^3 = -3.23x^2 + 5.54x - 9.84$$

$$x^3 - 3.23x^2 - 5.54x + 9.84 = 0$$

Нарисуем графики функции:

$$y = x^3$$

и 

$$z = -3.23x^2 + 5.54x - 9.84$$

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0.02, 0.2, 20)

y = np.x**3
z = -3.23*x**2 + 5.54*x - 9.84

plt.xlabel('x')
plt.grid(True)
plt.plot(x, y, label='x**3')
plt.plot(x, z, label='-3.23*x**2 + 5.54*x - 9.84')
plt.legend(loc='upper right')
plt.show()
```
