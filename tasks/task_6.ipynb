### 1. Метод последовательного поиска



$$x^3 - 75 = 0$$



Нарисуем графики функции:

$$y = x^3$$

и 

$$z = 75$$

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(2, 5, 20)

y = x**3
z = 75*np.ones(len(y))

plt.xlabel('x')
plt.grid(True)
plt.plot(x, y, label='x^3')
plt.plot(x,z, label='75')
plt.legend(loc='upper right')
plt.show()
```

Пусть:

$$f(x) = x^3 - 75$$
