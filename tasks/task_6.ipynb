### 1. Метод последовательного поиска

##### Пример:

$$x^3 = 3.23x^2 + 5.54x - 9.84$$

$$x^3 - 3.23x^2 - 5.54x + 9.84 = 0$$

Нарисуем графики функции:

$$y = x^3$$

и 

$$z = 3.23x^2 + 5.54x - 9.84$$

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-4, 0, 20)

y = x**3
z = 3.23*x**2 + 5.54*x - 9.84

plt.xlabel('x')
plt.grid(True)
plt.plot(x, y, label='x^3')
plt.plot(x, z, label='3.23*x^2 + 5.54*x - 9.84')
plt.legend(loc='upper right')
plt.show()
```

Пусть:

$$f(x) = x^3 - 3.23x^2 - 5.54x + 9.84$$

```python

from modules.equations.single import root_search

def f(x: float): return x**3 - 3.23*x**2 - 5.54*x + 9.84

x1, x2 = root_search(f, -4, 0, 20)

x1, x2
```

```python
f(x1), f(x2)
```
