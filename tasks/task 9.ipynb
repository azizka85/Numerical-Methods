### Пример

$y = \sqrt{\frac{x^2+x+3}{2+x}}$, где $x \in [-\frac{1}{2}, \frac{1}{2}]$

```python
import numpy as np
import random as rnd
from modules.fitting.poly_fit import poly_fit, std_dev, plot_poly

xd = np.linspace(-1./2., 1./2., 20)
yd = xd * np.sqrt((xd**2+xd+3)/(2+xd))

for i in range(len(yd)):
    yd[i] = yd[i] + yd[i] * rnd.random() * 0.1

c = poly_fit(xd, yd, 3)

plot_poly(xd, yd, c)
```

```python
std_dev(c, xd, yd)
```
