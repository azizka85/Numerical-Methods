### Пример №1

```python
import numpy as np
import matplotlib.pyplot as plt
import modules.interpolation.neville as nv
import modules.interpolation.cubic_spline as cs

xd = np.array([-4, -2, -1, 1, 3, 4])
yd = np.array([-53, -1, 4, 2, 24, 59])

x = np.linspace(-4, 4, 20)

n = len(x)

y = np.zeros((n, 2))
k = cs.curvatures(xd, yd)

for i in range(n):
    y[i, 0] = nv.polynom(xd, yd, x[i])
    y[i, 1] = cs.polynom(xd, yd, k, x[i])

plt.plot(xd, yd, 'o', x, y[:, 0], '-', x, y[:, 1], '--')
plt.xlabel('x')
plt.legend(['Data', 'Neville', 'Cubic spline'], loc='lower right')
```

### Пример №2

$y = \sqrt{x^2-x+9}*cos^2(x)+\sqrt{5-x}*sin(x)$, где $x \in [\frac{-\pi}{2}, \frac{\pi}{3}]$

```python
from math import pi, cos, sin
xd = np.linspace(-pi/2., pi/3., 5)
yd = np.sqrt(xd**2 - xd + 9)*np.cos(xd)**2 + np.sqrt(5 - xd)*np.sin(xd)

x = np.linspace(-pi / 2., pi / 3., 20)

n = len(x)

y = np.zeros((n, 2))
k = cs.curvatures(xd, yd)

for i in range(n):
    y[i, 0] = nv.polynom(xd, yd, x[i])
    y[i, 1] = cs.polynom(xd, yd, k, x[i])

plt.plot(xd, yd, 'o', x, y[:, 0], '-', x, y[:, 1], '--')
plt.xlabel('x')
plt.legend(['Data', 'Neville', 'Cubic spline'], loc='lower right')
```
