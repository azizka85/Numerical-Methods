```python
import numpy as np
import matplotlib.pyplot as plt
import modules.interpolation.neville as nv
import modules.interpolation.cubic_spline as cs

xd = np.array([0., 0.5, 1., 1.5, 2.])
yd = np.array([-0.7854, 0.6529, 1.7390, 2.2071, 1.9425])

x = np.linspace(0., 2., 20)

n = len(x)

y = np.zeros((n, 2))
k = cs.curvatures(xd, yd)

for i in range(n):
    y[i, 0] = nv.polynom(xd, yd, x[i])
    y[i, 1] = cs.polynom(xd, yd, k, x[i])

plt.plot(xd, yd, 'o', x, y[:, 0], '-', x, y[:, 1], '--')
plt.xlabel('x')
plt.legend(['Data', 'Neville', 'Cubic spline'], loc='lower right')
```

###  №2

$y = \sqrt{x^2-x+3}*cos(x)+\sqrt{2-x}$, где $x \in [\frac{-\pi}{2}, \frac{\pi}{3}]$

```python
from math import pi, cos
xd = np.linspace(-pi / 2., pi / 3., 5)
yd = np.sqrt(xd**2 - xd +3) * np.cos(xd) + np.sqrt(2 - xd)

x = np.linspace(-pi / 2., pi / 3., 20)

n = len(x)

y = np.zeros((n, 2))
k = cs.curvatures(xd, yd)

for i in range(n):
    y[i, 0] = nv.polynom(xd, yd, x[i])
    y[i, 1] = cs.polynom(xd, yd, k, x[i])

plt.plot(xd, yd, 'o', x, y[:, 0], '-', x, y[:, 1], '--')
plt.xlabel('x')
plt.legend(['Data', 'Neville', 'Cubic spline'], loc='lower right')
```
