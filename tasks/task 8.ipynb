### Пример №1

```python
import numpy as np
import matplotlib.pyplot as plt
import modules.interpolation.neville as nv
import modules.interpolation.cubic_spline as cs

xd = np.array([0., 21.1, 37.8, 54.4, 71.1, 87.8, 100.])
yd = np.array([1.79, 1.13, 0.696, 0.519, 0.338, 0.321, 0.296])

x = np.linspace(0., 100., 20)

n = len(x)

y = np.zeros((n, 2))
k = cs.curvatures(xd, yd)

for i in range(n):
    y[i, 0] = nv.polynom(xd, yd, x[i])
    y[i, 1] = cs.polynom(xd, yd, k, x[i])

plt.plot(xd, yd, 'o', x, y[:, 0], '-', x, y[:, 1], '--')
plt.xlabel('x')
plt.legend(['Data', 'Neville', 'Cubic spline'], loc='lower right')
```

### Пример №2

$y = \sqrt{\frac{x^2+x+3}{2+x}}$, где $x \in [-\frac{1}{2}, \frac{1}{2}]$

```python
xd = np.linspace(-1/2., 1/2., 5)
yd = np.sqrt((xd**2+xd+3)/(2+xd))

x = np.linspace(-1/2., 1/2., 20)

n = len(x)

y = np.zeros((n, 2))
k = cs.curvatures(xd, yd)

for i in range(n):
    y[i, 0] = nv.polynom(xd, yd, x[i])
    y[i, 1] = cs.polynom(xd, yd, k, x[i])

plt.plot(xd, yd, 'o', x, y[:, 0], '-', x, y[:, 1], '--')
plt.xlabel('x')
plt.legend(['Data', 'Neville', 'Cubic spline'], loc='lower right')
```
