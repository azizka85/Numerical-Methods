# Погрешности вычислений

### 1. Абсолютная и относительная погрешность

Пусть $x$ - точное, но, как правило, неизвестное значение, а $\widetilde{x}$ - её известное приближенное значение.

Абсолютная погрешность: $\triangle x = |x - \widetilde{x}|$

Относительная погрешность: $\delta = \frac{\triangle x}{|\widetilde{x}|}$

```python
from math import pi

x_exact = 2.03
x_approx = 2

error = abs(x_exact - x_approx)
delta = error / abs(x_approx)

print(f'Точное значение: {x_exact}')
print(f'Приближенное значение: {x_approx}')
print(f'Абсолютная погрешность: {error}')
print(f'Относительная погрешность: {delta}')
print(f'Относительная погрешность (%): {delta * 100}')
```

### 2. Погрешность суммы или разности чисел

Пусть $\widetilde{x_1} = x_1 + \triangle x_1$, $\widetilde{x_2} = x_2 + \triangle x_2$

тогда $\widetilde{x_1} + \widetilde{x_2} = x_1 + x_2 + \triangle x_1 + \triangle x_2$, т.е. $\triangle(x_1 + x_2) = \triangle x_1 + \triangle x_2$ 

$\triangle(x_1 - x_2) = \triangle x_1 + \triangle x_2$ 

$\triangle(x_1 + x_2 + ... + x_n) = \triangle x_1 + \triangle x_2 + ... + \triangle x_n = n \cdot \triangle x$ 

$
    \delta(x_1 \pm x_2) = \frac{\triangle(\widetilde{x_1} \pm \widetilde{x_2})}{\widetilde{x_1} \pm \widetilde{x_2}} = 
    \frac{\triangle(\widetilde{x_1}) \pm \triangle{\widetilde{x_2}}}{\widetilde{x_1} \pm \widetilde{x_2}} =
    \frac{\widetilde{x_1}}{\widetilde{x_1} \pm \widetilde{x_2}} \delta(\widetilde{x_1}) + \frac{\widetilde{x_2}}{\widetilde{x_1} \pm \widetilde{x_2}} \delta(\widetilde{x_2})
$

Пусть $max(\delta(x_i)) = M$, $min(\delta(x_i)) = m$, тогда $m \leq \delta(x) \leq M$

```python
x1_exact = 2.03
x1_approx = 2

error1 = abs(x1_exact - x1_approx)
delta1 = error1 / abs(x1_approx)

print(f'Точное значение x1 = {x1_exact}')
print(f'Приближенное значение x1 = {x1_approx}')
print(f'Абсолютная погрешность x1 = {error1}')
print(f'Относительная погрешность x1 = {delta1}')
```

```python
x2_exact = 3.07
x2_approx = 3

error2 = abs(x2_exact - x2_approx)
delta2 = error2 / abs(x2_approx)

print(f'Точное значение x2 = {x2_exact}')
print(f'Приближенное значение x2 = {x2_approx}')
print(f'Абсолютная погрешность x2 = {error2}')
print(f'Относительная погрешность x2 = {delta2}')
```

```python
error = abs(x1_exact + x2_exact - x1_approx - x2_approx)
delta = error / abs(x1_approx + x2_approx)

print(f'Абсолютная погрешность суммы x1 + x2 = {error}')
print(f'Относительная погрешность суммы x1 + x2 = {delta}')
```

```python
error = abs(x1_exact + x2_exact - x1_approx - x2_approx)
delta = error / abs(x1_approx + x2_approx)

print(f'Абсолютная погрешность разности x1 - x2 = {error}')
print(f'Относительная погрешность разности x1 - x2 = {delta}')
```

### 3. Погрешность произведения и деления чисел

$\widetilde{x_1} \cdot \widetilde{x_2} = (x_1 + \triangle x_1) \cdot (x_2 + \triangle x_2) = x_1 \cdot x_2 + x_2 \cdot \triangle x_1 + x_1 \cdot \triangle x_2$, т.е. $\triangle(x_1 \cdot x_2) = x_2 \cdot \triangle x_1 + x_1 \cdot \triangle x_2$ 

$\delta(x_1 \cdot x_2) = \frac{\triangle(x_1 \cdot x_2)}{x_1 \cdot x_2} = \delta(x_1) + \delta(x_2)$

$\delta(x_1 \cdot x_2 \cdot ... \cdot x_n) = \delta(x_1) + \delta(x_2) + \delta(x_3) + ... + \delta(x_n)$

$\triangle(\frac{x_1}{x_2}) = \frac{x_1 \cdot \triangle(x_2) + x_2 \cdot \triangle(x_1)}{{x_2}^2}$

$\delta(x_1/x_2/.../x_n) = \delta(x_1) + \delta(x_2) + ... + \delta(x_n)$

```python
error = abs(x1_exact * x2_exact - x1_approx * x2_approx)
delta = error / abs(x1_approx * x2_approx)

print(f'Абсолютная погрешность произведения x1 * x2 = {error}')
print(f'Относительная погрешность произведения x1 * x2 = {delta}')
```

```python
error = abs(x1_exact / x2_exact - x1_approx / x2_approx)
delta = error / abs(x1_approx / x2_approx)

print(f'Абсолютная погрешность частного x1 / x2 = {error}')
print(f'Относительная погрешность частного x1 / x2 = {delta}')
```

### 4. Погрешность функции

$z = f(x_1, x_2, ..., x_n)$

$\triangle(z) = |\frac{\partial f}{\partial x_1}| \cdot \triangle(x_1) + |\frac{\partial f}{\partial x_2}| \cdot \triangle(x_2) + ... + |\frac{\partial f}{\partial x_n}| \cdot \triangle(x_n)$

$\delta(z) = |\frac{x_1}{f} \cdot \frac{\partial f}{\partial x_1}| \cdot \delta(x_1) + 
|\frac{x_2}{f} \cdot \frac{\partial f}{\partial x_2}| \cdot \delta(x_2) + ... +
|\frac{x_n}{f} \cdot \frac{\partial f}{\partial x_n}| \cdot \delta(x_n)$ 

Пример:

Дана функция: $f(x) = sin(x)$

Необходимо найти погрешность функции в точке: $x = \pi/6$, если приближенное значение $\pi = 3.141$

Тогда:

$\triangle(f) = cos(3.141/6) * (\pi/6 - 3.141/6)$

$\delta(f) = |\frac{3.141}{6} \cdot cos(3.141/6)| * \frac{\pi/6 - 3.141/6}{3.141/6}$

```python
from math import pi, sin

f_exact = sin(pi/6)
f_approx = sin(3.141/6)

error = abs(f_approx - f_exact)
delta = error / abs(f_approx)

print(f'Абсолютная погрешность: {error}')
print(f'Относительная погрешность: {delta}')
```
