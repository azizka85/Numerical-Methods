### Пример

$y= \sqrt{𝑥 + 1} − \sqrt{𝑥^2 + 19𝑥 + 123} $   где 𝑥 ∈ [0, 2]

##### Нахождение $F(x) = \int_1^x f(x)dx$ методом Ньютона-Котеса

```python
import numpy as np
from math import log
import modules.integration.newton_cotes as nc

a = 0
b = 2
k = 5

def f(x): return np.sqrt(𝑥 + 1) - np.sqrt(x**2 + 19*x +123) 

dx = [0.2, 0.02, 0.002]

x = [np.arange(a, b + d / 2, d) for d in dx]

i_nc = [np.zeros(len(t)) for t in x]

for i in range(len(x)):
    n = len(x[i])

    for j in range(n):
        i_nc[i][j] = nc.integral(f, a, x[i][j], k)        
```

```python
from matplotlib import pyplot as plt

for i in range(len(dx)):
    plt.plot(x[i], i_nc[i], label=f'dx = {dx[i]}')

plt.title("Интеграл методом Ньютона-Котеса")
plt.xlabel('x')
plt.grid()
plt.legend()
```

##### Нахождение $F(x) = \int_1^x f(x)dx$ методом Гаусса

```python
import modules.integration.gauss as g

i_g = [np.zeros(len(t)) for t in x]

for i in range(len(x)):
    n = len(x[i])

    for j in range(n):
        i_g[i][j] = g.integral(f, a, x[i][j], k)   
```

```python
for i in range(len(dx)):
    plt.plot(x[i], i_g[i], label=f'dx = {dx[i]}')

plt.title("Интеграл методом Гаусса")
plt.xlabel('x')
plt.grid()
plt.legend()
```

##### Сравнение $F(x) = \int_1^x f(x)dx$ для $dx = 0,2$

```python
plt.plot(x[0], i_nc[0], label=f'Метод Ньютона-Котеса')
plt.plot(x[0], i_g[0], label=f'Метод Гаусса')

plt.title("Сравнение интегралов для dx = 0,2")
plt.xlabel('x')
plt.grid()
plt.legend()
```
