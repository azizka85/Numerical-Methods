### 1. Метод последовательного поиска

##### Пример:

$$x^3 = 1.2x^2 + 8.19x - 13.23$$

$$x^3 - 1.2x^2 - 8.19x + 13.23 = 0$$

Нарисуем графики функции:

$$y = x^3$$

и 

$$z = 1.2x^2 + 8.19x - 13.23$$

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-5, -1, 20)

y = x**3
z = 1.2*x**2 + 8.19*x - 13.23

plt.xlabel('x')
plt.grid(True)
plt.plot(x, y, label='x**3')
plt.plot(x, z, label='1.2*x**2 + 8.19*x - 13.23')
plt.legend(loc='upper right')
plt.show()
```

Пусть:

$$f(x) = x^3 - 1.2x^2 - 8.19x + 13.23$$

```python
from modules.equations.single import root_search

def f(x: float): return x**3 - 1.2*x**2 - 8.19*x + 13.23

x1, x2 = root_search(f, -5, -1, 0.02)

x1, x2
```

```python
f(x1), f(x2)
```
