### 1. Метод последовательного поиска

##### Пример:

$$tg(x)-th(x)=0$$

Нарисуем графики функции:

$$y = tg(x)$$

и 

$$z = th(x)$$

```python
import numpy as np
from matplotlib import pyplot as plt


x = np.linspace(-2, -1.25, 20)

y = np.tan(x)
z = np.tanh(x)

plt.xlabel('x')
plt.grid(True)
plt.plot(x, y, label='tan(x)')
plt.plot(x, z, label='th(x)')
plt.legend(loc='upper right')
plt.show()
```

Пусть:

$$f(x)=tg(x)-th(x)$$

```python
from math import tan, tanh
from modules.equations.single import root_search

def f(x: float): return np.tan(x)-np.tanh(x)

x1, x2 = root_search(f, -2, -1.25, 0.001)

x1, x2
```

```python
f(x1), f(x2)
```
