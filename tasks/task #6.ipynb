### 1. Метод последовательного поиска

##### Пример:

$$ ch(x) * cos(x) - 1 = 0 $$


Нарисуем графики функции:

$$y=(exp(x)+exp(-x))/2*cos(x)-1$$
$$z=0$$


```python
import numpy as np
from math import pi
from matplotlib import pyplot as plt

x = np.linspace(0, 8, 20)

y = np.cos(x)
z =1

plt.xlabel('x')
plt.grid(True)
plt.plot(x, y, label='(exp(x)+exp(-x))/2*cos(x)')
plt.plot(x, z, label='0')
plt.legend(loc='upper right')
plt.show()
```

Пусть:

$$ch(x)*cos(x)-1=0$$
