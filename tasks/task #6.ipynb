### 1. Метод последовательного поиска

##### Пример:

$$x^2-2=0$$

Нарисуем графики функции:

$$y = x^2$$

и 

$$z = 2$$

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0, 2, 20 )

y = x**2
z = 2*np.ones(len(y))

plt.xlabel('x')
plt.grid(True)
plt.plot(x, y, label='ln(x)')
plt.plot(x, z, label='x - 2')
plt.legend(loc='upper right')
plt.show()
```
