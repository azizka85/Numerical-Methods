### 1. Метод последовательного поиска

##### Пример:

$$x^3-7x^2+9=0$$


Нарисуем графики функции:

$$y = x^3$$

и 

$$z = 7x^2 - 9$$

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-5., 0., 10)

y = x**3
z = 7*x**2 - 9

plt.xlabel('x')
plt.grid(True)
plt.plot(x, y, label='x^3')
plt.plot(x, z, label='7x^2 - 9')
plt.legend(loc='upper right')
plt.show()
```

Пусть:

$$f(x) = x^3 - 7x^2 + 9$$

```python
from modules.equations.single import root_search

def f(x: float): return x**3.0 - 7.0*x**2 + 9.0

x1, x2 = root_search(f, -5.0, 0, 0.02)

x1, x2
```

```python
f(x1), f(x2)
```
