$$x^4+2x^3-7x^2+3=0$$

Нарисуем графики функции

$$y=x^4$$

$$z=-2X^3+7X^2-3$$

### 1. Метод последовательного поиска

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-1, 1, 2)

y = x**4
z = -2*x**3 + 7*x**2 - 3

plt.xlabel('x')
plt.grid(True)
plt.plot(x, y, label='x^4')
plt.plot(x, z, label='-2*X**3+7*X**2-3')
plt.legend(loc='upper right')
plt.show()
```
