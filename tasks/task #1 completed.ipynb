### 1. Абсолютная и относительная погрешность

Пусть $x$ - точное, но, как правило, неизвестное значение, а $\widetilde{x}$ - её известное приближенное значение.

Абсолютная погрешность: $\triangle x = |x - \widetilde{x}|$

Относительная погрешность: $\delta = \frac{\triangle x}{|\widetilde{x}|}$

```python

a = 3.05
b = 3

error = abs(a - b)
delta = error / abs(a)

print(f'Точное значение: {a}')
print(f'Приближенное значение: {b}')
print(f'Абсолютная погрешность: {error}:')
print(f'Относительная погрешность: {delta}')
print(f'Относительная погрешность (%): {delta * 100}')
```

### 2. Погрешность суммы или разности чисел

Пусть $\widetilde{x_1} = x_1 + \triangle x_1$, $\widetilde{x_2} = x_2 + \triangle x_2$

тогда $\widetilde{x_1} + \widetilde{x_2} = x_1 + x_2 + \triangle x_1 + \triangle x_2$, т.е. $\triangle(x_1 + x_2) = \triangle x_1 + \triangle x_2$ 

$\triangle(x_1 - x_2) = \triangle x_1 + \triangle x_2$ 

$\triangle(x_1 + x_2 + ... + x_n) = \triangle x_1 + \triangle x_2 + ... + \triangle x_n = n \cdot \triangle x$ 

$
    \delta(x_1 \pm x_2) = \frac{\triangle(\widetilde{x_1} \pm \widetilde{x_2})}{\widetilde{x_1} \pm \widetilde{x_2}} = 
    \frac{\triangle(\widetilde{x_1}) \pm \triangle{\widetilde{x_2}}}{\widetilde{x_1} \pm \widetilde{x_2}} =
    \frac{\widetilde{x_1}}{\widetilde{x_1} \pm \widetilde{x_2}} \delta(\widetilde{x_1}) + \frac{\widetilde{x_2}}{\widetilde{x_1} \pm \widetilde{x_2}} \delta(\widetilde{x_2})
$

Пусть $max(\delta(x_i)) = M$, $min(\delta(x_i)) = m$, тогда $m \leq \delta(x) \leq M$

```python
a = 3.05
b = 3
c = 3.04
d = 3

e = a + c
f = b + d

error = abs(e - f)
delta = error / abs(e)

print("Для a + c")
print(f'Точное значение: {e}')
print(f'Приближенное значение: {f}')
print(f'Абсолютная погрешность: {error}:')
print(f'Относительная погрешность: {delta}')
print(f'Относительная погрешность (%): {delta * 100}')
print("-----------------------------------------------")

e = a - c
f = b - d

error = abs(e - f)
delta = error / abs(e)
print("Для a - c")
print(f'Точное значение: {e}')
print(f'Приближенное значение: {f}')
print(f'Абсолютная погрешность: {error}:')
print(f'Относительная погрешность: {delta}')
print(f'Относительная погрешность (%): {delta * 100}')
print("-----------------------------------------------")

```
