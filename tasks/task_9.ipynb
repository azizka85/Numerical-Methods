```python
import numpy as np
import random as rnd
from modules.fitting.poly_fit import poly_fit, std_dev, plot_poly

xd = np.linspace(-pi / 2., pi / 3., 35)
yd = np.sqrt(xd**2 - xd +3) * np.cos(xd) + np.sqrt(2 - xd)

for i in range(len(yd)):
    yd[i] = yd[i] + yd[i] * rnd.random() * 0.125

c = poly_fit(xd, yd, 1)

plot_poly(xd, yd, c)
```
