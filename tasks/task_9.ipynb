### Пример

$y = \sqrt{x^2-x+9}*cos^2(x)+\sqrt{5-x}*sin(x)$, где $x \in [\frac{-\pi}{2}, \frac{\pi}{3}]$

```python
from math import pi
import numpy as np
import random as rnd
from modules.fitting.poly_fit import poly_fit, std_dev, plot_poly

xd = np.linspace(-pi / 2., pi / 3., 15)
yd = np.sqrt(xd**2-xd+9)*np.cos(xd)**2+np.sqrt(5-xd)*np.sin(xd)

for i in range(len(yd)):
    yd[i] = yd[i] + yd[i] * rnd.random() * 0.1

c = poly_fit(xd, yd, 3)

plot_poly(xd, yd, c)
```

```python
std_dev(c, xd, yd)
```
