### Метод Ньютона-Рафсона для системы уравнении

##### Пример:

$$2y^2 + xy - 12 = 0$$

$$x^2 + 4xy + 20 = 0$$


```python
import numpy as np
from math import sin, log
from modules.equations.system import newton_raphson_system

def f(x: np.ndarray[np.double]) -> np.ndarray[np.double]:
    fx = np.zeros(len(x))

    fx[0] = 2*x[1]**2 + x[0]*x[1] - 12
    fx[1] = x[0]**2 + 4*x[0]*x[1] + 20.

    return fx

x = np.array([1., 1.])

x, iter = newton_raphson_system(f, x, 10**-4, 10**-9, 100)

x, iter
```
