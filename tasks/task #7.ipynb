### 5. Метод Ньютона-Рафсона для системы уравнении

##### Пример:

$$2/x+3/y=8$$

$$5/x-2/y=1$$



```python
import numpy as np
from modules.equations.system import newton_raphson_system

def f(x: np.ndarray[np.double]) -> np.ndarray[np.double]:
    fx = np.zeros(len(x))

    fx[0] = 2 / x[0] + 3 / x[1] - 8
    fx[1] = 5 / x[0] - 2 / x[1] - 1

    return fx

x = np.array([11., 20.])
x, iter = newton_raphson_system(f, x, 10**-4, 10**-9, 100)

x, iter
```
