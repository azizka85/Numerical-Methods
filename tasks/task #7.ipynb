### 1. Метод Ньютона-Рафсона для системы уравнении

##### Пример:

$$x^2 - 3xy - 16= 0$$

$$3x^2 + 2xy - y^2 = 0$$



```python
import numpy as np
from math import sin, log
from modules.equations.system import newton_raphson_system

def f(x: np.ndarray[np.double]) -> np.ndarray[np.double]:
    fx = np.zeros(len(x))

    fx[0] = x[0]**2 - 3*x[0]*x[1] - 16.
    fx[1] = 3*x[0]**2 + 2*x[0]*x[1] - x[1]**2.


    return fx

x = np.array([1., 1.])

x, iter = newton_raphson_system(f, x, 10**-4, 10**-9, 100)

x, iter
```
