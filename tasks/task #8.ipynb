```python
import numpy as np
import matplotlib.pyplot as plt
import modules.interpolation.neville as nv
import modules.interpolation.cubic_spline as cs

xd = np.array([-2.0, -1.5, -0.1, 0.5])
yd = np.array([2.2796, 1.6467, 1.0025, 1.0635])

x = np.linspace(-2, 0.5, 20)

n = len(x)

y = np.zeros((n, 2))
k = cs.curvatures(xd, yd)

for i in range(n):
    y[i, 0] = nv.polynom(xd, yd, x[i])
    y[i, 1] = cs.polynom(xd, yd, k, x[i])

plt.plot(xd, yd, 'o', x, y[:, 0], '-', x, y[:, 1], '--')
plt.xlabel('x')
plt.legend(['Data', 'Neville', 'Cubic spline'], loc='lower right')
```

###  №2

$y=\sqrt{\frac{x^2-x+8}{8-x}}$  где $x \in [\frac{-1}{2}, \frac{1}{2}]$


```python
xd = np.linspace(-1/2., 1/2., 5)
yd = np.sqrt((xd**2 - xd + 8 )/ (8 - xd ))

x = np.linspace(-1/2., 1/2., 20)

n = len(x)

y = np.zeros((n, 2))
k = cs.curvatures(xd, yd)

for i in range(n):
    y[i, 0] = nv.polynom(xd, yd, x[i])
    y[i, 1] = cs.polynom(xd, yd, k, x[i])

plt.plot(xd, yd, 'o', x, y[:, 0], '-', x, y[:, 1], '--')
plt.xlabel('x')
plt.legend(['Data', 'Neville', 'Cubic spline'], loc='lower right')
```
