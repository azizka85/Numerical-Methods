### Пример №1

```python
import numpy as np
import matplotlib.pyplot as plt
import modules.interpolation.neville as nv
import modules.interpolation.cubic_spline as cs

xd = np.array([0.0, 0.0204, 0.1055, 0.241, 0.582, 0.712, 0.981])
yd = np.array([0.385, 1.04, 1.79, 2.63, 4.39, 4.99, 5.27])

x = np.linspace(0.0, 0.981, 20)

n = len(x)

y = np.zeros((n, 2))
k = cs.curvatures(xd, yd)

for i in range(n):
    y[i, 0] = nv.polynom(xd, yd, x[i])
    y[i, 1] = cs.polynom(xd, yd, k, x[i])

plt.plot(xd, yd, 'o', x, y[:, 0], '-', x, y[:, 1], '--')
plt.xlabel('x')
plt.legend(['Data', 'Neville', 'Cubic spline'], loc='lower right')
```

### Пример №2

$y = \sqrt{\frac{x^2-x+9}{5-x}}$, где $x \in [-\frac{1}{2}, \frac{1}{2}]$

```python
xd = np.linspace(-1/2., 1/2., 5)
yd = np.arcsinh(30*xd)

x = np.linspace(-1/2., 1/2., 20)

n = len(x)

y = np.zeros((n, 2))
k = cs.curvatures(xd, yd)

for i in range(n):
    y[i, 0] = nv.polynom(xd, yd, x[i])
    y[i, 1] = cs.polynom(xd, yd, k, x[i])

plt.plot(xd, yd, 'o', x, y[:, 0], '-', x, y[:, 1], '--')
plt.xlabel('x')
plt.legend(['Data', 'Neville', 'Cubic spline'], loc='lower right')
```
