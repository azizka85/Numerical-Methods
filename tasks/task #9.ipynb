### Пример

$y = | x \cdot sh(x) |$, где $x \in [-2, 2]$

```python
import numpy as np
import random as rnd
from modules.fitting.poly_fit import poly_fit, std_dev, plot_poly

xd = np.linspace(-2., 2., 20) 
yd = np.abs ( xd * np.sinh(xd) )

for i in range(len(yd)):
    yd[i] = yd[i] + yd[i] * rnd.random() * 0.11

c = poly_fit(xd, yd, 2)

plot_poly(xd, yd, c)
```

```python
std_dev(c, xd, yd)
```
